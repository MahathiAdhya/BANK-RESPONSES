```python
from ipywidgets import interact, interact_manual
```

```python
df.select_dtypes('object')
```

```python
df.select_dtypes('number').columns
```

```python
@interact_manual
def user_boxplot(col =df.select_dtypes('number').columns):
    sns.boxplot(df[col])
```

```python
@interact
def user_info(col =df.select_dtypes('number').columns):
    return df[col].describe()
```

```python
@interact_manual
def user_boxplot(col =df.select_dtypes('object').columns):
    sns.countplot(df[col])
```

```python
@interact_manual
def user_barplot(col =df.select_dtypes('object').columns,
                col1 =df.select_dtypes('number').columns):
    sns.barplot(df[col],df[col1])
```

```python
df.groupby(['marital','job_role'])['age'].agg({'min','max','mean','median','count'})
```

```python
@interact
def user_barplot(col =df.select_dtypes('object').columns,
                col1 =df.select_dtypes('number').columns,
                 col2 =df.select_dtypes('object').columns):
    return df.groupby([col,col2])[col1].agg({'min','max','mean','median','count'})
```

```python
d = []
for i in df['duration']:
    if 'sec' in i:
        d.append(float(i.split()[0])/60)
    else:
        d.append(float(i.split()[0]))
```

```python
df['duration'] = d
```
